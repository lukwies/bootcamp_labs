Refer to the `files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv` dataset.

```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
```

#### 1. Get the numerical variables from our dataset.


```python
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
numerical = df.select_dtypes(np.number)
numerical.head()
```

#### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
def plot_distribution(data):
    for col in data.columns:
        sns.displot(x=data[col], height=2)
```

```python
plot_distribution(numerical)
```

##### So all of the previous plots show a exponential-like distribution

#### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
from sklearn.preprocessing import PowerTransformer

pT = PowerTransformer()
pT.fit(numerical)
numerical_trans = pd.DataFrame(pT.transform(numerical), columns=numerical.columns)
```

```python
plot_distribution(numerical_trans)
```

```python
from sklearn.preprocessing import StandardScaler

sS = StandardScaler()
sS.fit(numerical)
numerical_trans = pd.DataFrame(sS.transform(numerical), columns=numerical.columns)
```

```python
plot_distribution(numerical_trans)
```
